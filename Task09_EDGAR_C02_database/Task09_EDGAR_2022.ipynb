```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# EDGAR data reader to read sheet 1 and 4. The other sheets have different formats.
def read_EDGAR_sheet(filename, sheet):
    df_tmp=pd.read_excel(filename, sheet, usecols = "C:BC")
    df_tmp=df_tmp[~df_tmp["Country"].isna()]
    df_tmp = df_tmp.T
    headers = df_tmp.iloc[0]
    df = df_tmp.iloc[1:].copy()
    df.rename(headers, axis=1,inplace = True)
    return(df)
```

```python
datafile = r"https://edgar.jrc.ec.europa.eu/booklet/EDGARv7.0_FT2021_fossil_CO2_booklet_2022.xlsx"
# IN case you do not have internet: datafile = r"./data/EDGARv7.0_FT2021_fossil_CO2_booklet_2022.xlsx"
```

```python
# Read sheet 1
df1 = read_EDGAR_sheet(datafile, sheet=1)
```

```python
df1.columns[df1.columns.str.contains("N")]
```

```python
# Read sheet 4
df2 = read_EDGAR_sheet(datafile, sheet=4)
```

```python
fig1, (ax11, ax21, ax31) = plt.subplots(3,1,figsize=(10,18),dpi=136)

df1[["China", "United States", "India", "Germany","Kenya"]].plot(ax = ax11)
ax11.grid(True)
ax11.set_title("Title 1?")
ax11.set_xlabel("x-Label?")
ax11.set_ylabel("Quantity / Unit?")
ax11.grid(True)

df2[["China", "United States", "India", "Germany","Kenya"]].plot(ax = ax21)
ax21.grid(True)
ax21.set_title("Title 2?")
ax21.set_xlabel("x-Label?")
ax21.set_ylabel("Quantity / Unit?")
ax21.grid(True)

(df1/df2)[["China", "United States", "India", "Germany","Kenya"]].plot(ax = ax31)
ax31.grid(True)
ax31.set_title("Title 3?")
ax31.set_xlabel("x-Label?")
ax31.set_ylabel("Quantity?")

plt.show()
```
